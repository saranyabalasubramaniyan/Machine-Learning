That will make our feature engineering efforts easier later on.

```python
#Displays the columns in the data frame(from the csv)
print("Columns from the ufo Sighting Data: ")
print(ufoData.columns)
print()
#Displays the data types of the columns in the data frame
print("Data types of columns from the ufo Data Sightings:")
print(ufoData.dtypes)

#Converting the date column to Date time type, if there are errors NaN will be assigned
ufoData["date"] =  pd.to_datetime(ufoData["date"],errors='coerce')
#Converting the city column to string
ufoData["city"] = ufoData["city"].astype('string')
#Converting the state column to string
ufoData["state"] = ufoData["state"].astype('string')
#Converting the country column to string
ufoData["country"] = ufoData["country"].astype('string')
#Converting the type column to string
ufoData["type"] = ufoData["type"].astype('string')
#Converting the length_of_time column to string
ufoData["length_of_time"] = ufoData["length_of_time"].astype('string')
#Converting the desc column to string
ufoData["desc"] = ufoData["desc"].astype('string')
#Converting the recorded column to string, if there are errors NaN will be assigned
ufoData["recorded"] = pd.to_datetime(ufoData["recorded"],errors='coerce')
#Converting the recorded column to numeric, if there are errors NaN will be assigned
ufoData["lat"] = pd.to_numeric(ufoData["lat"],errors='coerce')

#Displaying the data types of the columns after the type conversion
print()
print("Data types of columns from the ufo Data Sightings after type conversion: ")
print(ufoData.dtypes)
```

## 4. Extracting numbers from strings (10 points)
The <b>length_of_time</b> column in the UFO dataset is a text field that has the number of 
minutes within the string. 
Here, you'll extract that number from that text field using regular expressions.

```python
import re
def return_minutes(time_string):
    
    # Use \d+ to grab digits
    pattern = re.compile(r"\d+")
     ############################################Consider any value with a number in it
    # Use match on the pattern and column
    num = re.match(pattern, time_string)
    
    if num is not None:
        """ 1 week = 7 days
        1 day = 24 hours
        1 hour = 60 minutes
        1 day = 24 * 60 minutes
        1 week = 24 * 60 * 7 minutes
        1 week = 10080 minutes"""
        if 'week' in time_string:
            return (int(num.group(0))*10080)
        """1 day = 24 hours
        1 hour = 60 minutes
        1 day = 24 * 60 minutes
        1 day = 1440 minutes"""
        if 'hour' in time_string:
            return (int(num.group(0))*1440)
        """1 hour = 60 minutes"""
        if 'hour' in time_string:
            return (int(num.group(0))*60)
        """ 1 minute = 60 seconds
        1 second = 0.0167 seconds
        """
        if 'sec' in time_string:
            return (int(num.group(0))*0.0167)
        if ':' in time_string:
            timeList = time_string.split(":")
            if len(timeList) == 3:
                """1 hour = 60 minutes
                 1 second = 0.0167 seconds"""
                try:
                    return int(timeList[0])*60+int(timeList[1])+int(timeList[2])*0.0167
                except:
                     return int(num.group(0))
        return int(num.group(0))
         
# Apply the extraction to the length_of_time column
ufoData_CLEANED["length_of_time_minutes"] = ufoData_CLEANED["length_of_time"].apply(return_minutes)
 
# Take a look at the head of both of the columns
display(ufoData_CLEANED[['length_of_time','length_of_time_minutes']].head())
print("Current size of the dataframe : " + str(ufoData_CLEANED.shape))
```

```python
import numpy as npy
print("Variance of seconds column : " + str(ufoData_CLEANED["seconds"].var()))
print("Variance of length_of_time_minutes column : " + str(ufoData_CLEANED["length_of_time_minutes"].var()))
print("Variance of lat column : " + str(ufoData_CLEANED["lat"].var()))
print("Variance of long column : " + str(ufoData_CLEANED["long"].var()))
print()

#With the results, high variance is observed in seconds column
# Log normalize the seconds column
ufoData_CLEANED["seconds_log"] = npy.log(ufoData_CLEANED['seconds'])
#Log Normalize length_of_time_minutes column
ufoData_CLEANED["lot_mins_log"] = npy.log(ufoData_CLEANED[['length_of_time_minutes']])
#Log Normalize lat column
ufoData_CLEANED["lat_log"] = npy.log(ufoData_CLEANED[['lat']])
#Log Normalize long column
ufoData_CLEANED["long_log"] = npy.log(ufoData_CLEANED[['long']])

print("-------------- Evaluating the seconds column ------------------")
#Displaying variance after log normalization
print("Variance after log normalizing the seconds column: " + str(ufoData_CLEANED['seconds_log'].var()))
print("Variance is NaN because the data contains invalid values like 0. Hence filtering the seconds column for 0")
#Filtering invalid values in the seconds column
ufoData_NORM = ufoData_CLEANED.loc[ufoData_CLEANED['seconds'] > 0] 
print("Variance after the cleanup of invalid values : " + str(ufoData_NORM["seconds"].var()))
#Log normalizing seconds column after the processing for invalid values
ufoData_CLEANED["seconds_norm"] = npy.log(ufoData_NORM['seconds'])
#Displaying variance after log normalization
print("Variance after log normalizing the seconds column: " + str(ufoData_CLEANED["seconds_norm"].var()))
print(ufoData_NORM.shape)
print()

print("-------------- Evaluating the length_of_time_minutes column ------------------")
#Displaying variance after log normalization
print("Variance after log normalizing the length_of_time_minutes column: " + str(ufoData_CLEANED['lot_mins_log'].var()))
print("Variance is NaN because the data contains invalid values like 0. Hence filtering the length_of_time_minutes column for 0")
#Filtering invalid values in the length_of_time_minutes column
ufoData_NORM = ufoData_CLEANED.loc[ufoData_CLEANED['length_of_time_minutes'] > 0] 
print("Variance after the cleanup of invalid values : " + str(ufoData_NORM["length_of_time_minutes"].var()))
#Log normalizing seconds column after the processing for invalid values
ufoData_CLEANED["lot_mins_norm"] = npy.log(ufoData_NORM['length_of_time_minutes'])
#Displaying variance after log normalization
print("Variance after log normalizing the seconds column: " + str(ufoData_CLEANED["lot_mins_norm"].var()))
print(ufoData_NORM.shape)
print()
```
